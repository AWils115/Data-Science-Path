## Data Preparation
We now need to prepare the data for the model. We first need to decide what variables we are going to include in the model. Since we are wanting to predict Zodiac signs, we need to pick variables we feel are relevant to this. Therefore the variables we will use are: 'body_type', 'diet', 'orientation', 'pets', 'religion', 'sex' and 'job'.

ML algorithms don't react well to missing values so lets see if there are any in the variables we have chosen. We also need to clean the data in 'religion' because, similar to signs, a lot of the answers have extra wording that isn't required.

```python
profiles['religion_cleaned'] = profiles['religion'].str.split().str.get(0)
profiles[['body_type', 'diet', 'orientation', 'pets', 'religion_cleaned', 'sex', 'job', 'sign_cleaned']].isnull().sum()
```

#### Drop NA
We can see there are a lot of null answers in the variables we have selected. We will need to create a new `DataFrame` with these null values taken out. Dropping the na values still gives us a dataset with over 16,000 datapoints, enough for our model.

```python
cols = ['body_type', 'diet', 'orientation', 'pets', 'religion_cleaned', 'sex', 'job', 'sign_cleaned']
df = profiles[cols].dropna()
df.shape
```

#### Dummy Variables
In order for the algorithm to function correctly we need to create dummy variables for the categorical variables. This will split each categorical variable into each possible answer and give a binary result for each one.

```python
for col in cols[:-1]:
    df = pd.get_dummies(df, columns=[col], prefix=[col])
df.head()
```

#### Label Imbalance
An imbalance in the prediction label needs to be checked. This is important since it's a multi-class problem where two or more outcomes can be had. An imbalance in a response variable is bad since it means that some labels only occur a few times. This is an issue for machine learning algorithms if there are not enough data to train with which will give bad predictions.

In the given dataset, we observe that the counts of all the zodiac signs are more or less equal (i.e., without large deviations). Hence, we do not have to worry about imbalances and trying to address this problem.

```python
df.sign_cleaned.value_counts()
```

#### Splitting Data
Next the data needs to be split into train and validation sets. In this split 25% of the data is reserved for the final validation, while 75% is kept for training the model.

```python
col_length = len(df.columns)

# Splitting the data into X and Y, Y is the target column (col 1) and X is the rest
X = df.iloc[:, 1:col_length]
Y = df.iloc[:, 0:1]

# The size of our validation set is 25% of the data
val_size = 0.25

# Split the data
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=val_size, random_state=0)

# Turn into 1D arrays
Y_train = Y_train.to_numpy().ravel()
Y_val = Y_val.to_numpy().ravel()
```

## Prediction

#### Model Building
We can now build some models with the data, for this project we will use Logistic Regression, Decision Trees and KNN.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

#### Logisitic Regression
The first model will be Logistic Regression with the `multi_class='multinomial'` argument. 

```python
lr_model = LogisticRegression(multi_class='multinomial')
lr_model.fit(X_train, Y_train)
lr_predictions = lr_model.predict(X_train)
print(classification_report(Y_train, lr_predictions))
```

We can see above that the model was accurate only 12% of the time. This isn't good as a random guess would result in being right 1/12 = 8% of the time.

#### K Nearest Neighbor
The next model will be K Nearest Neighbor which will take 20 of its neighbrs to predict the signs. The default value of 5 was kept for `n_neighbors`.

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, Y_train)
knn_predicitons = knn_model.predict(X_train)
print(classification_report(Y_train, knn_predicitons))
```

This model was accurate 33% of the time which is a step in the right direction.

#### Decision Trees
The last model is a decision tree, with the default `max_depth=none` which means that it will "If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples."

```python
cart_model = DecisionTreeClassifier()
cart_model.fit(X_train, Y_train)
cart_predicitons = cart_model.predict(X_train)
print(classification_report(Y_train, cart_predicitons))
```

This model was much better, being accurate 78% of the time. 

#### Confusion Matrix
Below is a confusion matrix of the decision tree model. This shows the true labels on the y-axis and the predicted labels on the x-axis. The diagnol represents the true positives, because we can see higher numbers and lighter colors here, it confirms our models accuracy at predicting zodiac signs.

```python
# Generate confusion_matrix
from sklearn.metrics import confusion_matrix
cart_cm = confusion_matrix(Y_train, cart_predicitons)
cart_labels = cart_model.classes_

# Plot confusion_matrix as heatmap
f, ax = plt.subplots(figsize=(15, 8))

sns.heatmap(cart_cm, annot=True, ax=ax, fmt='d')
ax.set_ylabel('True Labels')
ax.set_xlabel('Predicted Labels')
ax.xaxis.set_ticklabels(cart_labels)
ax.yaxis.set_ticklabels(cart_labels)
ax.yaxis.set_tick_params(rotation=360)
ax.xaxis.set_tick_params(rotation=90)
```

```python
cart_model.get_depth()
```

The decision tree model has a max depth of 49 branches. This suggests the model has been overfit to the data and will not generalize across other datasets.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

kfold = KFold(n_splits=5, shuffle=True, random_state=0)
results = cross_val_score(cart_model, X_train, Y_train, cv=kfold, scoring='accuracy')

print(results)
print('Baseline: %.2f%% (%.2f%%)' % (results.mean()*100, results.std()*100))
```

Using a 5 fold cross validation we get a baseline of around 9% which is worse than the KNN and about the same as logistic regression.

We will now build the decision tree again but limit the `max_depth` to 20, to stop the algorithm reaching a stopping point.

```python
cart_model20 = DecisionTreeClassifier(max_depth=20)
cart_model20.fit(X_train, Y_train)
cart_predictions20 = cart_model20.predict(X_train)

print(classification_report(Y_train, cart_predictions20))
```

```python
results20 = cross_val_score(cart_model20, X_train, Y_train, cv=kfold, scoring='accuracy')

print(results20)
print('Baseline: %.2f%% (%.2f%%)' % (results20.mean()*100, results.std()*100))
```

We can see that by limiting the max depth the accuracy has come right down to 41% and the cross validation is around 8.5%.
