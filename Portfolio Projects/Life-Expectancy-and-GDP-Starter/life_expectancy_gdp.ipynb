# Life Expectancy and GDP

## What is the average life expectancy and GDP?

```python
means = data.drop('Year', axis=1).groupby('Country').mean().reset_index()
means
```

```python
medians = data.drop('Year', axis=1).groupby('Country').median().reset_index()
medians
```

We now have the mean and median for each nation. Again we can see that there is an outlier with Zimbabwe, their life expectancy and GDP is much lower than the rest of the nations. To better compare the data lets create some plots.

```python
sns.set_style('darkgrid')
sns.set_palette('pastel')
sns.set_context('poster', font_scale=0.5)
```

```python
figure, axes = plt.subplots(1, 2, figsize=(15, 8), sharey=True)
axes[0] = sns.barplot(data=means, x='GDP', y='Country', ax=axes[0])
axes[0].set_xlabel('GDP (Trillion Dollars)')
axes[0].set_title('Mean GDP')

axes[1] = sns.barplot(data=means, x='Life Expectancy', y='Country', ax=axes[1])
axes[1].set_xlabel('Life Expectancy (years)')
axes[1].set_title('Mean Life Expectancy')
plt.show()
```

```python
figure, axes = plt.subplots(1, 2, figsize=(15, 8), sharey=True)
axes[0] = sns.barplot(data=medians, x='GDP', y='Country', ax=axes[0])
axes[0].set_xlabel('GDP (Trillion Dollars)')
axes[0].set_title('Median GDP')

axes[1] = sns.barplot(data=medians, x='Life Expectancy', y='Country', ax=axes[1])
axes[1].set_xlabel('Life Expectancy (years)')
axes[1].set_title('Median Life Expectancy')
plt.show()
```

Now we've created these plots we have a better visual representation of the data. We can see now that although at first we thought Zimbabwe was an outlier for being so low, though true, we can now see that the GDP figure for the United States is much larger than for the other nations and so could be considered an outlier itself. Another interesting point is that despite such a large GDP difference the life expectancy for the United States isn't actually the largest of the nations, which is Germany. We can also see that despite Chile having a relatively low GDP the life expectancy here is on par with China, Mexico, Germany and USA. 

## Has GDP increased over time?

```python
plt.figure(figsize=(15, 8))
GDP = sns.lineplot(data=data, x='Year', y='GDP', hue='Country', legend='full')
GDP.set_title('GDP Change Over Time for each Country')
GDP.legend(fontsize=15)
GDP.set_ylabel('GDP (Trillion Dollars)')
plt.show()
```

As we predicted earlier, because of the large difference in GDP between Zimbabwe, Chile and the other nations it is difficult to properly discern the changes in GDP over time for all the nations on one graph. We need to plot seperate graphs for each country to explore in detail.

```python
plt.figure(figsize=(15, 8))
GDPgrid = sns.FacetGrid(data, col='Country', col_wrap=3, hue='Country', sharey=False)
GDPgrid.map(sns.lineplot, 'Year', 'GDP').add_legend().set_axis_labels('Year', 'GDP (Trillion Dollars)')
GDPgrid.set_titles(' ')
plt.show()
```

We now have a much clearer view of whats going on. Despite some years of lower GDP, generally the GDP of each nation has incereased throughout the years from 2000 to 2015. China and USA have had a much smoother increases in GDP over the years compared to the other nations. Zimbabwe from 2000 to around 2008 had consecutive years of lower GDP figures before suddenly taking off and quadrupling their GDP figure up to 2015. Chile, Germany and Mexico have struggled to maintain their trend in the later years. 

We are now able to answer our question, despite some bumps, all six nations have increased their GDP figures from 2000 to 2015.

## Has life expectancy increased over time?

```python
plt.figure(figsize=(15, 8))
LE = sns.lineplot(data=data, x='Year', y='Life Expectancy', hue='Country', legend='full')
LE.set_title('Life Expectancy Change Over Time for each Country')
LE.legend(fontsize=15, loc='lower right')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
LEgrid = sns.FacetGrid(data, col='Country', col_wrap=3, hue='Country', sharey=False)
LEgrid.map(sns.lineplot, 'Year', 'Life Expectancy').add_legend().set_axis_labels('Year', 'Life Expectancy (years)')
LEgrid.set_titles(' ')
plt.show()
```

We can see that generally life expectancy has increased over time for each of the six nations. The lines for life expectancy do look similar to the GDP lines we plotted earlier, which may give us an early indication of how they are correlated. China and USA again have a smoother increase in life expectancy, whereas the other nations have more of a choppy increase. Zimbabwe also has a dip in life expectancy up to 2005 before a constant increase to over 60. 

## Are GDP and life expectancy correlated?

```python
plt.figure(figsize=(15, 8))
GDPLEscatter = sns.scatterplot(y='GDP', x='Life Expectancy', hue='Country', data=data)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
GDPLEgrid = sns.FacetGrid(data, col='Country', col_wrap=3, hue='Country', sharey=False, sharex=False)
GDPLEgrid.map(sns.scatterplot, 'Life Expectancy', 'GDP').add_legend().set_axis_labels('Life Expectancy (years)', 'GDP (Trillion Dollars)')
GDPLEgrid.set_titles(' ')
plt.show()
```

We can see that life expectancy and GDP do have a strong positive correlation for each nation. Some nations have a stronger correlation than others and we could attribute that to other effects.

## What is the distribution of life expectancy and GDP?

```python
figure, axes = plt.subplots(1, 2, figsize=(15, 8), sharey=True)
axes[0] = sns.violinplot(data=data, x='GDP', y='Country', ax=axes[0])
axes[0].set_xlabel('GDP (Trillion Dollars)')
axes[0].set_title('Distribution of GDP')

axes[1] = sns.violinplot(data=data, x='Life Expectancy', y='Country', ax=axes[1])
axes[1].set_xlabel('Life Expectancy (years)')
axes[1].set_title('Distribution of Life Expectancy')
plt.show()
```

From these plots we can see how the distribution varies between nations. For GDP: Zimbabwe, Chile and Mexico have much tighter ranges, showing that relative to China, USA and Germany their GDP growth has been as signifcant. For the life expectancy: Zimbabwe has a much larger range, shwoing they have come a long way to increase the life expectancy of their poplulation.
