# Roller Coaster
This project is from the Codecademy Data Science career path. The objective is to create several data visualizations that will give an insight into the world of roller coasters.

Data for the project is contained within 3 csv files: Golden_Ticket_Award_Winners_Wood.csv, Golden_Ticket_Award_Winners_Steel.csv and roller_coasters.csv. We are going to focus on Golden_Ticket_Award_Winners_Wood.csv and Golden_Ticket_Award_Winners_Steel.csv first, so I will read the csv's and create dataframes for them.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wood = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
steel = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')
```

These two dataframes contain information for roller coasters made of steel or wood, that are ranked for the Golden Ticket awards from 2013-2018. We are going to create a function that will plot the ranking of a given roller coaster over time.

```python
def plot_ranking(data, coaster_name, park_name):
    max_rank = wood.Rank[(wood.Name == coaster_name) & (wood.Park == park_name)].max()
    ax = plt.subplot()
    plt.plot(data['Year of Rank'][(data.Name == coaster_name) & (data.Park == park_name)],
             data.Rank[(data.Name == coaster_name) & (data.Park == park_name)], 'o-', label=coaster_name)
    ax.invert_yaxis()
    ax.set_xticks([2013, 2014, 2015, 2016, 2017, 2018])
    ax.set_yticks(range(1, max_rank + 1))
    plt.title(f'{coaster_name} of {park_name} Park Ranking from 2013 to 2018')
    plt.ylabel('Ranking')
    plt.xlabel('Year')
    plt.legend()
    plt.show()
```

```python
plot_ranking(wood, 'El Toro', 'Six Flags Great Adventure')
```

Next we want a function that will plot the rankings of two rollercoasters over time.

```python
def plot_two_rankings(data, coaster1_name, coaster2_name, coaster1_park, coaster2_park):
    coasters_max_rank = [data.Rank[(data.Name == coaster1_name) & (data.Park == coaster1_park)].max(), 
                         data.Rank[(data.Name == coaster2_name) & (data.Park == coaster2_park)].max()]
    ax = plt.subplot()
    plt.plot(data['Year of Rank'][(data.Name == coaster1_name) & (data.Park == coaster1_park)],
             data.Rank[(data.Name == coaster1_name) & (data.Park == coaster1_park)], 'o-', label=coaster1_name)
    plt.plot(data['Year of Rank'][(data.Name == coaster2_name) & (data.Park == coaster2_park)],
             data.Rank[(data.Name == coaster2_name) & (data.Park == coaster2_park)], 's-', label=coaster2_name)
    ax.invert_yaxis()
    ax.set_yticks(range(1, max(coasters_max_rank) + 1))
    ax.set_xticks([2013, 2014, 2015, 2016, 2017, 2018])
    plt.title(f'{coaster1_name} vs {coaster2_name} Rankings 2013 to 2018')
    plt.ylabel('Ranking')
    plt.xlabel('Year')
    plt.legend()
    plt.show()
```

```python
plot_two_rankings(wood, 'El Toro', 'Boulder Dash', 'Six Flags Great Adventure', 'Lake Compounce')
```

Next we want a function that will plot the rankings of the top n roller coasters over time.

```python
def top_n_plot(data, n):
    top_n_coasters = data[data.Rank <= n]
    for coaster in top_n_coasters.Name:
        ax = plt.subplot()
        plt.plot(top_n_coasters['Year of Rank'][top_n_coasters.Name == coaster], 
                 top_n_coasters.Rank[top_n_coasters.Name == coaster], 'o-', label=coaster)
        ax.invert_yaxis()
        ax.set_yticks(range(1, n + 1))
        ax.set_xticks([2013, 2014, 2015, 2016, 2017, 2018])
        plt.ylabel('Rank')
        plt.xlabel('Year')
        plt.legend()
        plt.show()        
```

```python
top_n_plot(wood, 5)
```

```python
coasters = pd.read_csv('roller_coasters.csv')
coasters.head()
```

```python
def plot_hist(data, col_name):
    if col_name == 'height':
        plt.hist(data[col_name][data[col_name] <= 140])
    else:
        plt.hist(data[col_name])
    plt.xlabel(col_name)
    plt.ylabel('Count')
    plt.show()
```

```python
plot_hist(coasters, 'height')
```

```python
def no_inversions(data, park_name):
    ax = plt.subplot()
    plt_data = data[data['park'] == park_name]
    plt.bar(plt_data['name'], plt_data['num_inversions'])
    ax.set_xticks(range(len(plt_data['name'])))
    ax.set_xticklabels(plt_data['name'], rotation=30)
    plt.xlabel('Coaster Name')
    plt.ylabel('No. Inversions')
    plt.show()
```

```python
no_inversions(coasters, 'Parc Asterix')
```

```python
def plot_pie(data):
    status_counts = [len(data[data.status == 'status.operating']), len(data[data.status == 'status.closed.definitely'])]
    plt.pie(status_counts, autopct='%0.1f%%', labels=['Operating','Closed'])
    plt.show()
```

```python
plot_pie(coasters)
```

```python
def plot_scatter(data, column_x, column_y):
    if column_x == 'height':
        plt.scatter(data[column_x][data[column_x] <= 140], data[column_y][data[column_x] <= 140])
    elif column_y == 'height':
        plt.scatter(data[column_x][data[column_y] <= 140], data[column_y][data[column_y] <= 140])
    else:    
        plt.scatter(data[column_x], data[column_y])
    plt.xlabel(column_x)
    plt.ylabel(column_y)
    plt.show()
```

```python
plot_scatter(coasters, 'height', 'speed')
```
