#### Linear Regression, Single Feature-

We will now focus on creating a model to predict the success of a tennis player. First we will create a model that takes only one feature from the data. This feature will be Aces.

```python
# Loading data for model:

aces = df[['Aces']]
winnings = df[['Winnings']]
```

```python
# Split the data into test and training

aces_train, aces_test, winnings_train, winnings_test = train_test_split(aces, winnings, train_size=0.8)
```

```python
# Training the model

model = LinearRegression()
model.fit(aces_train, winnings_train)
```

```python
# Scoring the model
print(f'Predicting winnings based on Aces, test score: {round(model.score(aces_test, winnings_test), 3)}')
```

```python
# Make predictions

winnings_predict = model.predict(aces_test)

plt.figure(figsize=(15, 8))
plt.scatter(winnings_test, winnings_predict, alpha=0.4)
plt.title('Predicted Winnings vs Actual Winnings - 1 Feature')
plt.xlabel('Winnings Test Data')
plt.ylabel('Winnings Predicted Data')
plt.show()
```

Using the one feature Aces has created a good model of our winnings, with a test score of 0.595, meaning around 60% of the predicted winnings can be explained by the feature Aces.

We will now create another model using one feature, this feature will be first serve.

```python
# Loading data for model

firstserve = df[['FirstServe']]
winnings = df[['Winnings']]

# Split data to test and training

firstserve_train, firstserve_test, winnings_train, winnings_test = train_test_split(firstserve, 
                                                                                          winnings, train_size=0.8)

# Training the model

model = LinearRegression()
model.fit(firstserve_train, winnings_train)

# Scoring the model

print(f'Predicting winnings based on first serve, test score: {round(model.score(firstserve_test, winnings_test), 3)}')
```

```python
# Make predictions

winnings_predict = model.predict(firstserve_test)

plt.figure(figsize=(15, 8))
plt.scatter(winnings_test, winnings_predict, alpha=0.4)
plt.title('Predicted Winnings vs Actual Winnings - 1 Feature')
plt.xlabel('Winnings Test Data')
plt.ylabel('Winnings Predicted Data')
plt.show()
```

We can see that the x-scale and y-scale differ by quite a bit and so it looks like using first serve is not a good feature for creating the model, this is backed up by the test score of -0.003, meaning the predicted winnings are only 0.3% explained by first serve and it has a negative impact.

#### Linear Regression, Two Feature - 

We will now create a model based on two features from the dataset, these two features will be first serve points won and break points saved.

```python
# Loading data for model

features = df[['FirstServePointsWon', 'BreakPointsSaved']]
winnings = df[['Winnings']]

# Split data to test and training

features_train, features_test, winnings_train, winnings_test = train_test_split(features, winnings, train_size=0.8)

# Training the model

model = LinearRegression()
model.fit(features_train, winnings_train)

# Score the model

print(f'Predicting winnings based on first serve points won and break points saved, test score: {round(model.score(features_test, winnings_test), 3)}')
```

```python
# Make Predictions

winnings_predict = model.predict(features_test)

plt.figure(figsize=(15, 8))
plt.scatter(winnings_test, winnings_predict, alpha=0.4)
plt.title('Predicted Winnings vs Actual Winnings - 2 Features')
plt.xlabel('Winnings Test Data')
plt.ylabel('Winnings Predicted Data')
plt.show()
```

We can see again that the scales of the graph are different and so the predicted winnings don't match to the test data, this is again backed up by our test score of 0.159, meaning the predicted winnings are only around 16% explained by first serve points won and break points saved.

#### Linear Regression, Multiple Features - 

We will now create a model that uses multiple features to improve the test score and get more accurate predictions of winnings. The features I will use will be: FirstServe, FirstServePointsWon, FirstServeReturnPointsWon, SecondServePointsWon, SecondServeReturnPointsWon, Aces, BreakPointsConverted, BreakPointsFaced, BreakPointsOpportunities, BreakPointsSaved, DoubleFaults, ReturnGamesWon, ReturnPointsWon, ServiceGamesWon, TotalPointsWon, TotalServicePointsWon, Wins, Losses.

```python
df.columns
```

```python
# Load the data

features = df[['FirstServe', 'FirstServePointsWon', 'FirstServeReturnPointsWon', 'SecondServePointsWon',
             'SecondServeReturnPointsWon', 'Aces', 'BreakPointsConverted', 'BreakPointsFaced',
             'BreakPointsOpportunities', 'BreakPointsSaved', 'DoubleFaults', 'ReturnGamesWon', 'ReturnPointsWon',
             'ServiceGamesWon', 'TotalPointsWon', 'TotalServicePointsWon', 'Wins', 'Losses']]
winnings = df[['Winnings']]

# Split the data to train and test

features_train, features_test, winnings_train, winnings_test = train_test_split(features, winnings, train_size=0.8)

# Train the model

model = LinearRegression()
model.fit(features_train, winnings_train)

# Score the model

print(f'Predicting winnings based on multiple features, test score: {round(model.score(features_test, winnings_test), 3)}')
```

```python
# Make predictions

winnings_predict = model.predict(features_test)

plt.figure(figsize=(15, 8))
plt.scatter(winnings_test, winnings_predict, alpha=0.4)
plt.title('Actual vs Predicted Winnings - Multiple Features')
plt.xlabel('Winnings Test Data')
plt.ylabel('Winnings Predicted Data')
plt.show()
```

By using multiple features we have now been able to create a fairly accurate model, with a test score of 0.873. Our model is now able to use the features we have input to explain 87% of the predicted winnings.
